# 2D and 3D

```python
import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(-1* np.pi, np.pi,0.01)# -pi to pi
y = np.sin(x)
plt.plot(x,y,'g')
plt.show()
```

```python
 
x = np.arange(-1* np.pi, np.pi,0.3)# -pi to pi
y = np.sin(x)
plt.plot(x,y,'ro--') # marker 
plt.title('y = sin(x)',{'fontsize':20,'color':'w'})
plt.grid()
plt.legend(['Sin(x)'])
plt.xlabel('x')
plt.show()
```

#Semi-Log and Log-Log 2D plot

```python
x = [10**x for x in range(0,5)]
y = [10**y for y in range(0,5)]
plt.plot(x,y)
plt.show()
```

```python
plt.semilogx(x,y)
plt.show()
```

```python
plt.semilogy(x,y)
plt.show()
```

```python
plt.loglog(x,y)
plt.show()
```

```python
x= np.arange(-2*np.pi,2*np.pi,0.01)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,x,y2)
plt.legend(["sinx","$cos(x)$"])
plt.show()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["$sin(x)$","$cos(x)$"])
plt.show()
```

#Subplots

```python
x= np.arange(-2*np.pi,2*np.pi,0.01)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.subplot(1,3,1) #no of row, no of column,  index
plt.plot(x,y1)
plt.subplot(1,3,2) 
plt.plot(x,y2)
plt.subplot(1,3,3) 
plt.plot(x,y3)
plt.grid()
plt.show()
```

```python
plt.subplot(3,1,1) #no of row, no of column,  index
plt.plot(x,y1)
plt.subplot(3,1,2) 
plt.plot(x,y2)
plt.subplot(3,1,3) 
plt.plot(x,y3)
plt.show()
```
