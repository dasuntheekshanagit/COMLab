# 2D and 3D

```python
import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(-1* np.pi, np.pi,0.01)# -pi to pi
y = np.sin(x)
plt.plot(x,y,'g')
plt.show()
```

```python
 
x = np.arange(-1* np.pi, np.pi,0.3)# -pi to pi
y = np.sin(x)
plt.plot(x,y,'ro--') # marker 
plt.title('y = sin(x)',{'fontsize':20,'color':'w'})
plt.grid()
plt.legend(['Sin(x)'])
plt.xlabel('x')
plt.show()
```
